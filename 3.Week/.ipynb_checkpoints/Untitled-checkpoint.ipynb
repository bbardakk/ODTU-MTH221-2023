```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import Perceptron

# Load dataset
iris = datasets.load_iris()
X = iris.data[:100, :2]  # Using only two features for visualization
y = iris.target[:100]
```

```python
plt.scatter(X[:50, 0], X[:50, 1], color='red', marker='o', label='setosa')
plt.scatter(X[50:100, 0], X[50:100, 1], color='blue', marker='x', label='versicolor')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.legend(loc='upper left')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
```

```python
param_grid = {'eta0': [0.0001, 0.001, 0.01, 0.1, 1.0]}
perceptron = Perceptron(max_iter=1000, tol=1e-3, random_state=1)
grid_search = GridSearchCV(perceptron, param_grid, cv=5)
grid_search.fit(X_train, y_train)
print("Best parameters: {}".format(grid_search.best_params_))
```

```python
best_eta = grid_search.best_params_['eta0']

# Train with best parameter
clf = Perceptron(max_iter=1, eta0=best_eta, random_state=1, verbose=0, warm_start=True, tol=None)
clf.fit(X_train, y_train)

# For visualization, let's extract weights after first iteration
w1 = clf.coef_

# Train to convergence
clf.set_params(max_iter=1000, tol=1e-3)
clf.fit(X_train, y_train)

# Extract weights after final iteration
w2 = clf.coef_

# Visualize decision boundaries

fig, ax = plt.subplots(1, 2, figsize=(15, 5))
ax[0].scatter(X_train[:, 0], X_train[:, 1], c=y_train)
ax[0].set_title("Decision boundary after 1 iteration")
xx = np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1)
yy = -w1[0][0] / w1[0][1] * xx - (clf.intercept_[0]) / w1[0][1]
ax[0].plot(xx, yy, 'k--')

ax[1].scatter(X_train[:, 0], X_train[:, 1], c=y_train)
ax[1].set_title("Decision boundary after convergence")
yy = -w2[0][0] / w2[0][1] * xx - (clf.intercept_[0]) / w2[0][1]
ax[1].plot(xx, yy, 'k--')
plt.show()
```
